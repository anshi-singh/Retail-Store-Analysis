# Exploratory Data Analysis

```python
df['Sub-Category'].value_counts()
```

```python
plt.figure(figsize=(14,12))
df['Sub-Category'].value_counts().plot.pie()
plt.show()
```

# Observation:

Maximum sub-categories are from Binders, Paper, Furnishings, Phone, Storage, Art, Accessories and Minimum are from Machines, Copiers

```python
Profit_plot = (ggplot(df, aes(x='Sub-Category', y='Profit', fill='Sub-Category')) + geom_col() + coord_flip()
+ scale_fill_brewer(type='div', palette="Spectral") + theme_classic() + ggtitle('Profit Chart'))

display(Profit_plot)
```

# Observation:

The above chart shows the profit and loss of every subcategories.
Here from the graph we can visualize that "binders" sub-category has suffered the highest amount of loss and also profit amongst all other sub-Categories (For now we can't say that what is the reason it may be because of discounts given on binders subcategory).
Next,"Copiers" sub-category has gained highest amount of profit with no loss.There are other sub-categories too haven't faced any kind of losses but their profit margins are also low.
Next, suffering from highest loss is machines

```python
df['State'].value_counts()
```

```python
plt.figure(figsize=(15,12))
sns.countplot(x='State',data=df,order=df['State'].value_counts().index)
plt.xticks(rotation=90)
plt.show()
```

# Observation:

Highest number of buyers are from California, New York and Texas

```python
states_PS = df.groupby('State')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
states_PS[:].plot.bar(color=['pink','blue'],figsize=(15,8))
plt.title('Profit/loss & Sales across states')
plt.xlabel('States')
plt.ylabel('Profit/loss & Sales')
plt.show()
```

# Observation:

Profit is high for California, New york

Loss is faced in Texas, Pennsylvania, Ohio

```python
ps = df.groupby('Sub-Category')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
ps[:].plot.bar(color=['purple','blue'],figsize=(15,10))
plt.title('Profit/loss & Sales across states')
plt.xlabel('Sub-Category')
plt.ylabel('Profit/loss & Sales')
plt.show()
```
